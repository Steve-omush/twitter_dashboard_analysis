```python
import pandas as pd

# List of file names
file_names = ["C:/Users/MACTECH/Downloads/Compressed/dataset/kaggle/worldcup/worldcup.xlsx",
              "C:/Users/MACTECH/Downloads/Compressed/dataset/kaggle/ukraine/ukraine.csv",
              "C:/Users/MACTECH/Downloads/Compressed/dataset/kaggle/Bitcoin/bit.csv",
              "C:/Users/MACTECH/Downloads/Compressed/dataset/kaggle/covid/covid.xlsx",
              "C:/Users/MACTECH/Downloads/Compressed/dataset/kaggle/worldcup/worldcup-1.xlsx"
             ]

# Initialize an empty list to store dataframes
dfs = []

# Iterate over each file
for file_name in file_names:
    # Read the CSV or Excel file into a DataFrame
    if file_name.endswith('.csv'):
        df = pd.read_csv(file_name)
    elif file_name.endswith('.xlsx'):
        df = pd.read_excel(file_name, engine='openpyxl')
    else:
        print(f"Unsupported file format: {file_name}")
        continue
    
    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames in the list along rows
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
import pandas as pd
import re
import numpy as np


# Define a function to extract hashtags from a given text
def extract_hashtags(text):
    hashtags = re.findall(r"#(\w+)", text)
    return hashtags if hashtags else None

# Apply the hashtag extraction function to the "Tweet Content" column
sampled_df["Hashtags"] = sampled_df["Tweet Content"].apply(extract_hashtags)

# Replace 0 values in the "Hashtags" column with NaN
sampled_df["Hashtags"] = sampled_df["Hashtags"].replace(0, np.nan)

# Drop rows where the "Hashtags" column is missing
sampled_df = sampled_df.dropna(subset=["Hashtags"])

# Flatten the list of hashtags and count their frequency
hashtags = sampled_df["Hashtags"].explode().value_counts()

# Print the top 10 most popular hashtags
print(hashtags.head(10))
```
